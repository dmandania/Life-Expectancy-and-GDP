# **Life Expectancy Versus GDP**

## Introduction

The goal of this project is to analyse the data we have on the GDP over a 15 year period for the 6 countires in our database.

Generally speaking, we would assume that life expectancy would increase as GDP does. This is because GDP is a measure of how affluent a country is, with the more wealthier countries usually having access to better medical care.

However, we must also acknowledge that time is also a seperate factor that can contribute to an increase in life expectancy in an increasingly connected world. This is because scientific data and advancements are constantly occuring, and are generally freely shared via research papers and the internet so that any country in the world can benefit from them. These could be general studies on how to increase your lifespan by eating certain food, or doing a certain amount/type of physical activity every day. These sorts of effects are obviously independent of the GDP of any one individual country. There are also the continuous implementation of better health and safety standards around the world, with new laws to better protect consumers and workers coming in to effect over time.

## Preliminary Analysis

Let us start by importing the data and preparing it to be used.

```python
# Import all the libraries we should need
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Read in the csv file
gdp_data = pd.read_csv("all_data.csv")
print(gdp_data.head())

# Check which countries are in our data
countries = set(gdp_data["Country"])
print(countries)

# Save each countries data seperately so it is easier to fetch later
mexico = []
chile = []
zimbabwe = []
usa = []
china = []
germany = []

for i in range(len(gdp_data)):
    if gdp_data.Country[i] == "Mexico":
            mexico.append(gdp_data.iloc[i])
            i += 1
    elif gdp_data.Country[i] == "Chile":
            chile.append(gdp_data.iloc[i])
            i += 1
    elif gdp_data.Country[i] == "Zimbabwe":
            zimbabwe.append(gdp_data.iloc[i])
            i += 1
    elif gdp_data.Country[i] == "United States of America":
            usa.append(gdp_data.iloc[i])
            i += 1
    elif gdp_data.Country[i] == "China":
            china.append(gdp_data.iloc[i])
            i += 1
    elif gdp_data.Country[i] == "Germany":
            germany.append(gdp_data.iloc[i])
            i += 1
```

Now let us create an initial exploratory analysis of GDP against life expectancy for each country in our dataset.

```python
#Set up a 2x3 figure to group the charts together
fig, ax1 = plt.subplots(2, 3)
fig.set_size_inches(15, 10)

# Bar chart for Mexico
mexico_df = pd.DataFrame(mexico)
ax2 = ax1[0, 0].twinx()
ax1[0, 0].bar(mexico_df.Year, mexico_df.GDP, label = "GDP")
ax2.plot(mexico_df.Year, mexico_df["Life expectancy at birth (years)"], label = "Life expectancy", color = "red")
plt.title("Mexico")
plt.ylim([40, 85])
ax1[0, 0].set_xlabel("Year")
ax1[0, 0].set_ylabel("GDP (in Trillions)")
ax2.set_ylabel("Life expectancy (years)")
plt.legend(loc = "upper left")

# Bar chart for Chile
chile_df = pd.DataFrame(chile)
ax2 = ax1[0, 1].twinx()
ax1[0, 1].bar(chile_df.Year, chile_df.GDP, label = "GDP")
ax2.plot(chile_df.Year, chile_df["Life expectancy at birth (years)"], label = "Life expectancy", color = "red")
plt.title("Chile")
plt.ylim([40, 85])
ax1[0, 1].set_xlabel("Year")
ax1[0, 1].set_ylabel("GDP (in Trillions)")
ax2.set_ylabel("Life expectancy (years)")
plt.legend(loc = "upper left")

# Bar chart for Zimbabwe
zimbabwe_df = pd.DataFrame(zimbabwe)
ax2 = ax1[0, 2].twinx()
ax1[0, 2].bar(zimbabwe_df.Year, zimbabwe_df.GDP, label = "GDP")
ax2.plot(zimbabwe_df.Year, zimbabwe_df["Life expectancy at birth (years)"], label = "Life expectancy", color = "red")
plt.title("Zimbabwe")
plt.ylim([40, 85])
ax1[0, 2].set_xlabel("Year")
ax1[0, 2].set_ylabel("GDP (in Trillions)")
ax2.set_ylabel("Life expectancy (years)")
plt.legend(loc = "upper left")

# Bar chart for United States of America
usa_df = pd.DataFrame(usa)
ax2 = ax1[1, 0].twinx()
ax1[1, 0].bar(usa_df.Year, usa_df.GDP, label = "GDP")
ax2.plot(usa_df.Year, usa_df["Life expectancy at birth (years)"], label = "Life expectancy", color = "red")
plt.title("United States of America")
plt.ylim([40, 85])
ax1[1, 0].set_xlabel("Year")
ax1[1, 0].set_ylabel("GDP (in Trillions)")
ax2.set_ylabel("Life expectancy (years)")
plt.legend(loc = "upper left")

# Bar chart for China
china_df = pd.DataFrame(china)
ax2 = ax1[1, 1].twinx()
ax1[1, 1].bar(china_df.Year, china_df.GDP, label = "GDP")
ax2.plot(china_df.Year, china_df["Life expectancy at birth (years)"], label = "Life expectancy", color = "red")
plt.title("China")
plt.ylim([40, 85])
ax1[1, 1].set_xlabel("Year")
ax1[1, 1].set_ylabel("GDP (in Trillions)")
ax2.set_ylabel("Life expectancy (years)")
plt.legend(loc = "upper left")

# Bar chart for Germany
germany_df = pd.DataFrame(germany)
ax2 = ax1[1, 2].twinx()
ax1[1, 2].bar(germany_df.Year, germany_df.GDP, label = "GDP")
ax2.plot(germany_df.Year, germany_df["Life expectancy at birth (years)"], label = "Life expectancy", color = "red")
plt.title("Germany")
plt.ylim([40, 85])
ax1[1, 2].set_xlabel("Year")
ax1[1, 2].set_ylabel("GDP (in Trillions)")
ax2.set_ylabel("Life expectancy (years)")
plt.legend(loc = "upper left")
fig.tight_layout()
plt.show()
```

The above figure allows us to easily see certain trends within our dataset. The clearest trends are that both GDP and life expectancy tend to increase over time. However, we are able to pick up on some outliers that don't fit into this general trend.

For instance, Germany's GDP seems to be more or less flat from 2007 to 2015. However, its average life expectancy has still increased linearly during this time (even though it is by a small amount). This lends credence to our earlier hypothesis that life expectancy increases gradually over time due to advances in healthcare and medicine.

The clearest trend that I can see is that rate of increase (gradient) for life expectancy is more or less the same for all of the countires in our dataset (with the exception of Zimbabwe). This seems to be irrespective of little or how greatly their GDP has increased over the same period of time. For example, Chile's GDP has almost grown by 5 times and China's by 10 times, yet their rate of increase in life expectancy is still very similiar to that the USA or Germany which have seen a comparitively smaller increase in their GDPs. This is further evidence of our initial hypothesis that life expectancy has a greatly correlation to healthcare and medical advancements than to GDP.

Finally I would like to take a look at Zimbabwe, which seems to be the most obvious outlier in our dataset. This is the only country in our dataset which sees a substantial decrease in its GDP for the first few years of data. This corresponds with a decrease in average life expectancy too, which is counter to every other country that only ever increased their life expectancy. However, an important observation is that the average life expectancy starts to increase dramatically well before we see an increase in GDP. We can attribute the greater rate of increase to the fact that Zimbabwe has a significantly lower life expectancy to start with.

Therefore, we can confidently conclude that there is no significant correlation between GDP and average life expectancy.
